### Ejercicio 5

Añadid una nueva columna al _dataframe_ con un valor booleano indicando una predicción básica de si en una cierta batalla el bando atacante será vencedor. Definiremos el valor como _True_ sí el bando atacante es mayor al defensor, y como _False_ en caso contrario. Definid como _NaN_ aquellos casos en los que no haya número de tropas en ningún bando. ¿En que casos, identificados por el número de batalla (*battle\_number*), nuestra predicción coincide con el resultado real?

```python
# Respuesta
got_data_backup = got_data # creo una copia del DF para no perder datos
got_data_backup


```

```python
got_data["Prediccion"] = 0 # Creo la variable predicción para incluir los datos. 
for i in range(0,len(got_data)): # hago un for por todo el DF que devuelve los valores indicados. 
    if (pd.isna(got_data["attacker_size"][i]) == True) & (got_data["defender_size"][i]>0): # he usado el pd.isna para que cuando coincide un valor NaN con un valor numerico lo detecte y me devuelva True si el valor numerico lo tiene el atacante y false si lo tiene el defensor. 
        got_data["Prediccion"][i] = False
    elif (pd.isna(got_data["defender_size"][i]) == True) & (got_data["attacker_size"][i]>0):
        got_data["Prediccion"][i] = True
    elif (pd.isna(got_data["defender_size"][i]) == True) & (pd.isna(got_data["attacker_size"][i]) == True):
        got_data["Prediccion"][i] = "NaN"
    elif got_data["attacker_size"][i] < got_data["defender_size"][i]:
        got_data["Prediccion"][i] = False
    elif got_data["attacker_size"][i] > got_data["defender_size"][i]:
        got_data["Prediccion"][i] = True


got_data
```

```python
# Segunda parte del ejercicio
for i in range(0,len(got_data)): # hago otro for que compare las dos columnas y que imprima cuando los valores coincidan win-True y loss-false
    if (got_data["attacker_outcome"][i] == "win") & (got_data["Prediccion"][i] == True):
        print(got_data["battle_number"][i])
    elif (got_data["attacker_outcome"][i] == "loss") & (got_data["Prediccion"][i] == False):
        print(got_data["battle_number"][i])
```
